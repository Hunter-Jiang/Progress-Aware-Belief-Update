```python
save_idx = []
for js in s:
    goal = js["conversations"][2]["value"].split("Goal:")[-1].strip()[6:-1]
    #print(goal)
    try:
        idx = sdi[goal]
    except:
        idx = -1
    save_idx.append(idx)

sorted(save_idx)
#save_idx
```
